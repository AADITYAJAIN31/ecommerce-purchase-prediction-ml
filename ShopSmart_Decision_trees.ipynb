```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("shop_smart_ecommerce.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
class_count=df["Revenue"].value_counts()
plt.pie(class_count,labels=["False","True"],autopct="%1.1f%%")
plt.title("Total Revenue")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Weekend"]=le.fit_transform(df["Weekend"])
df["Revenue"]=le.fit_transform(df["Revenue"])
df.head()
```

```python
X=df.drop("Revenue",axis=1)
Y=df["Revenue"]
X=pd.get_dummies(X,columns=["Month","VisitorType"],drop_first=False,dtype=int)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
model=DecisionTreeClassifier(random_state=42)
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
print("accuracy score", accuracy_score(Y_test, y_pred))
print("precision score", precision_score(Y_test, y_pred))
print("recall score", recall_score(Y_test, y_pred))
print("f1 score",f1_score(Y_test,y_pred))
print("confusion matrix",confusion_matrix(Y_test,y_pred))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(18,20))
plot_tree(model,feature_names=X.columns,class_names=["No Purchase","Purchase"],max_depth=1,filled=True)
plt.tight_layout()
plt.show()
```

```python
max_depth=[2,3,4,5,6,7,8,9,10]
for depth in max_depth:
    model=DecisionTreeClassifier(class_weight="balanced",max_depth=depth)
    model.fit(X_train,Y_train)
    accuracy=model.score(X_test,Y_test)
    y_pred = model.predict(X_test) 
    F1=f1_score(Y_test,y_pred)
    print("for depth",depth,"accuracy =",accuracy,"f1 score",F1)
```

```python
min_sample_splits=[10,15,20,25,30]
for split in min_sample_splits:
    model=DecisionTreeClassifier(class_weight="balanced",max_depth=2,min_samples_split=split)
    model.fit(X_train,Y_train)
    accuracy=model.score(X_test,Y_test)
    y_pred=model.predict(X_test)
    F1=f1_score(Y_test,y_pred)
    print("for split",split,"accuracy =",accuracy,"f1 score",F1)
```

```python
full_tree=DecisionTreeClassifier(random_state=42)
full_tree.fit(X_train,Y_train)
```

```python
path=full_tree.cost_complexity_pruning_path(X_train,Y_train)
ccp_alphas=path.ccp_alphas
print(ccp_alphas)
```

```python
trees=[]
for alpha in ccp_alphas:
    model=DecisionTreeClassifier(random_state=42,ccp_alpha=alpha)
    model.fit(X_train,Y_train)
    trees.append((model,alpha))
```

```python
best_accuracy=0
best_alpha=0
for model,alpha in trees:
    curr_accuracy=model.score(X_test,Y_test)
    if curr_accuracy>best_accuracy:
        best_accuracy=curr_accuracy
        best_alpha=alpha
```

```python
print(best_accuracy)
print(best_alpha)
```

```python
best_model=DecisionTreeClassifier(ccp_alpha=best_alpha,max_depth=2)
best_model.fit(X_train,Y_train)
```

```python
y_pred=best_model.predict(X_test)
print("accuracy score", accuracy_score(Y_test, y_pred))
print("precision score", precision_score(Y_test, y_pred))
print("recall score", recall_score(Y_test, y_pred))
print("f1 score",f1_score(Y_test,y_pred))
print("confusion matrix",confusion_matrix(Y_test,y_pred))
```
